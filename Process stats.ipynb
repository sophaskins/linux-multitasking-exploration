```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

def make_graph(filename, stat, normalize=1, xlim=0, bins=None, labels={}, xlabel=None, ylabel=None):
    df = pd.read_csv("/home/sophaskins/datasets/" + filename)
    
    if not labels:
        for process_title in df.process.unique():
            labels[process_title] = process_title
            
    fig, ax = plt.subplots()
    fig.set_size_inches(14, 6)

    for process_title in df.process.unique():
        sns.distplot(df[df.process == process_title][stat] / normalize, label=labels[process_title], bins=bins)

    if xlim > 0:
        ax.set_xlim(0,xlim)
    if xlabel:
        ax.set(xlabel=xlabel)
    if ylabel:
        ax.set(ylabel=ylabel)

    plt.legend()
    plt.show()
```
